```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# dataset load
#load data
df = pd.read_csv("/content/drive/MyDrive/population/2015.csv")
df.info()
```

```python
df.head()
```

```python
# top 10 happoest countries
df.sort_values('Happiness Score' , ascending=False).head(10)[
    ['Country','Region','Happiness Score' ]
]

# this shows the countries with haighesst  happiness scores and majority are european countries
```

```python
# top 10 lowest  happiest countries

df.sort_values('Happiness Score' , ascending=True).head(10)[
    ['Country','Region','Happiness Score' ]

]
# the lowest happiest countries are in africa and  some asian countries
```

```python
region_happiness = df.groupby('Region')['Happiness Score'].mean().sort_values(ascending=False)
region_happiness
```

```python
# bar diagram of region and mean happiness
plt.bar(region_happiness.index, region_happiness.values)

plt.xticks(rotation=45, ha='right')
plt.xlabel('Region')
plt.ylabel('Avg happiness score')
plt.title("avg happiness score by region")
plt.show()
```

```python
# trust on goverment average score by region
trust_by_region = df.groupby('Region')['Trust (Government Corruption)'].mean().sort_values(ascending = False)
trust_by_region
```

```python
south_asia = df[df['Region'] =='Southern Asia'] [['Country','Freedom']].sort_values( 'Freedom',ascending=False)

south_asia
```

```python
GDP_vs_happiness= df.groupby("Region")['Happiness Score'].mean().sort_values(ascending = False)
GDP_vs_happiness
```

```python
happier_and_log_live =( df.sort_values(['Health (Life Expectancy)'],ascending = False).head(20)
[['Country','Health (Life Expectancy)','Happiness Score']
 ]
 )
happier_and_log_live
```

```python
df[['Health (Life Expectancy)', 'Happiness Score']].corr()
```

```python
# countries with high GDP but low happiness
gdp_mean = df['Economy (GDP per Capita)'].mean()
happiness_mean = df['Happiness Score'].mean()

high_gdp_low_happiness = df[
    (df['Economy (GDP per Capita)'] > gdp_mean) &
    (df['Happiness Score'] < happiness_mean)
][['Country', 'Economy (GDP per Capita)', 'Happiness Score']]

high_gdp_low_happiness
```

```python
import matplotlib.pyplot as plt

plt.scatter(df['Economy (GDP per Capita)'], df['Happiness Score'])
plt.xlabel('GDP per Capita')
plt.ylabel('Happiness Score')
plt.title('GDP vs Happiness')
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
plt.hist(df['Happiness Score'], bins=10)
plt.xlabel('Happiness Score')
plt.ylabel('Number of Countries')
plt.title('Global Distribution of Happiness Scores')
plt.show()
```

```python
# Factors that are strongly correleted with happiness
gdp_corr = df['Economy (GDP per Capita)'].corr(df['Happiness Score'])
family_corr = df['Family'].corr(df['Happiness Score'])
health_corr = df['Health (Life Expectancy)'].corr(df['Happiness Score'])
freedom_corr = df['Freedom'].corr(df['Happiness Score'])
trust_corr = df['Trust (Government Corruption)'].corr(df['Happiness Score'])
generosity_corr = df['Generosity'].corr(df['Happiness Score'])
dystopia_corr = df['Dystopia Residual'].corr(df['Happiness Score'])
```

```python
# bar comparison of these correlation

factors = [
    'GDP',
    'Family',
    'Health',
    'Freedom',
    'Trust',
    'Generosity',
    'Dystopia'
]

correlations = [
    gdp_corr,
    family_corr,
    health_corr,
    freedom_corr,
    trust_corr,
    generosity_corr,
    dystopia_corr
]

plt.bar(factors, correlations)
plt.xlabel('Factors')
plt.ylabel('Correlation with Happiness Score')
plt.title('Correlation of Factors with Happiness Score')
plt.xticks(rotation=45)
plt.show()
```

```python
# region and thier happiness S.D
region_variability = df.groupby('Region')['Happiness Score'].std()
region_variability
region_variability_sorted = region_variability.sort_values(ascending=False)
region_variability_sorted
```

```python
import matplotlib.pyplot as plt

plt.bar(region_variability_sorted.index, region_variability_sorted.values)
plt.xlabel('Region')
plt.ylabel('Standard Deviation of Happiness Score')
plt.title('Happiness Inequality Across Regions')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
plt.scatter(df['Happiness Rank'],df['Happiness Score'])
plt.xlabel('happiness rank')
plt.ylabel('happiness score')
plt.title('happiness ranks with thier score')
plt.show()
```

Summary: The analysis indicates that countries from regions such as Europe and AustraliaNew Zealand are generally more developed, with higher GDP per capita, longer life expectancy, and greater trust in government. However, several countries in South East Asia and Africa report comparatively high happiness scores despite lower economic and institutional indicators. This suggests that happiness does not depend solely on GDP, life expectancy, or government trust individually, but rather on a combination of social, cultural, and economic factors
